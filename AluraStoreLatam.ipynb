### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
total_precio = tienda["Precio"].sum()
total_precio2 = tienda2["Precio"].sum()
total_precio3 = tienda3["Precio"].sum()
total_precio4 = tienda4["Precio"].sum()
```

```python
print(f"El total del precio de la tienda 1 es: {total_precio}")
print(f"El total del precio de la tienda 2 es: {total_precio2}")
print(f"El total del precio de la tienda 3 es: {total_precio3}")
print(f"El total del precio de la tienda 4 es: {total_precio4}")
```

# 2. Ventas por categoría

```python

ventas_categoria = tienda.groupby("Categoría del Producto")["Categoría del Producto"].count().sort_values(ascending=False)
ventas_categoria2 = tienda2.groupby("Categoría del Producto")["Categoría del Producto"].count().sort_values(ascending=False)
ventas_categoria3 = tienda3.groupby("Categoría del Producto")["Categoría del Producto"].count().sort_values(ascending=False)
ventas_categoria4 = tienda4.groupby("Categoría del Producto")["Categoría del Producto"].count().sort_values(ascending=False)
```

```python
print(f"Las ventas por categoría de la tienda 1 son: \n{ventas_categoria}")
print(f"Las ventas por categoría de la tienda 2 son: \n{ventas_categoria2}")
print(f"Las ventas por categoría de la tienda 3 son: \n{ventas_categoria3}")
print(f"Las ventas por categoría de la tienda 4 son: \n{ventas_categoria4}")
```

# 3. Calificación promedio de la tienda


```python
calificacion_media = tienda["Calificación"].mean()
calificacion_media2 = tienda2["Calificación"].mean()
calificacion_media3 = tienda3["Calificación"].mean()
calificacion_media4 = tienda4["Calificación"].mean()
```

```python
print(f"La calificación media de la tienda 1 es: {calificacion_media}")
print(f"La calificación media de la tienda 2 es: {calificacion_media2}")
print(f"La calificación media de la tienda 3 es: {calificacion_media3}")
print(f"La calificación media de la tienda 4 es: {calificacion_media4}")
```

```python
# Tienda 1
print("Los 5 productos más vendidos de la tienda 1 es: ")
for producto, cantidad in productos_vendidos.head(5).items():
  print(f"{producto}: {cantidad}")
print("Los 5 productos menos vendidos de la tienda 1 es: ")
for producto, cantidad in productos_vendidos.tail(5).items():
  print(f"{producto}: {cantidad}")
print()

# Tienda 2
print("Los 5 productos más vendidos de la tienda 2 es: ")
for producto, cantidad in productos_vendidos2.head(5).items():
  print(f"{producto}: {cantidad}")

print("Los 5 productos menos vendidos de la tienda 2 es: ")
for producto, cantidad in productos_vendidos2.tail(5).items():
  print(f"{producto}: {cantidad}")
print()

# Tienda 3
print("Los 5 productos más vendidos de la tienda 3 es: ")
for producto, cantidad in productos_vendidos3.head(5).items():
  print(f"{producto}: {cantidad}")

print("Los 5 productos menos vendidos de la tienda 3 es: ")
for producto, cantidad in productos_vendidos3.tail(5).items():
  print(f"{producto}: {cantidad}")
print()

# Tienda 4
print("Los 5 productos más vendidos de la tienda 4 es: ")
for producto, cantidad in productos_vendidos4.head(5).items():
  print(f"{producto}: {cantidad}")

print("Los 5 productos menos vendidos de la tienda 4 es: ")
for producto, cantidad in productos_vendidos4.tail(5).items():
  print(f"{producto}: {cantidad}")
print()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
%matplotlib inline
```

```python
# Gráfico de barras: Ingresos totales por tienda
ingresos = {
    "Tienda 1": tienda["Precio"].sum(),
    "Tienda 2": tienda2["Precio"].sum(),
    "Tienda 3": tienda3["Precio"].sum(),
    "Tienda 4": tienda4["Precio"].sum()
}
ingresos_df = pd.Series(ingresos).sort_values(ascending=False)
plt.figure(figsize=(8,5))
sns.barplot(x=ingresos_df.index, y=ingresos_df.values, palette="Blues_d")
plt.title("Ingresos totales por tienda")
plt.ylabel("Ingreso total (suma de Precio)")
plt.xlabel("Tienda")
for i, v in enumerate(ingresos_df.values):
    plt.text(i, v, f"{v:.0f}", ha="center", va="bottom")
plt.tight_layout()
plt.show()
```

```python
# Gráfico de pastel: Distribución de categorías (todas las tiendas)
df_all = pd.concat([tienda.assign(Tienda='Tienda 1'),
                    tienda2.assign(Tienda='Tienda 2'),
                    tienda3.assign(Tienda='Tienda 3'),
                    tienda4.assign(Tienda='Tienda 4')], ignore_index=True)
cat_counts = df_all['Categoría del Producto'].value_counts()
top = cat_counts.head(8).copy()
others = cat_counts.iloc[8:].sum()
if others > 0:
    top['Otros'] = others
plt.figure(figsize=(7,7))
top.plot.pie(autopct='%1.1f%%', startangle=140, colors=plt.cm.tab20.colors)
plt.ylabel('')
plt.title('Distribución de categorías de producto (top 8 + Otros)')
plt.tight_layout()
plt.show()
```

```python
# Gráfico de dispersión: Precio vs Calificación (todas las tiendas)
sample = df_all.dropna(subset=['Precio', 'Calificación', 'Costo de envío'])
plt.figure(figsize=(8,6))
sc = plt.scatter(sample['Precio'], sample['Calificación'],
                 c=sample['Costo de envío'], cmap='viridis',
                 s=(sample['Costo de envío']+1)*10, alpha=0.6)
plt.colorbar(sc, label='Costo de envío')
plt.xlabel('Precio')
plt.ylabel('Calificación')
plt.title('Relación entre Precio y Calificación (todas las tiendas)')
plt.grid(True)
plt.tight_layout()
plt.show()
```
